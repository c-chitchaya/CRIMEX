```python
from sklearn_crfsuite import scorers,metrics
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_validate,train_test_split
from sklearn.metrics import precision_score
import sklearn_crfsuite
import numpy as np
```

```python
def doc2features(doc, i):
    word = doc[i][0]
    postag = doc[i][1]
    # Features from current word
#     {'word.word': 'DESC',
#    'word.isspace': False,
#    'postag': 'NCMN',
#    'word.isdigit()': False,
#    'word.prevword': '[::',
#    'word.previsspace': False,
#    'word.prepostag': 'NCMN',
#    'word.prevwordisdigit': False,
#    'word.nextword': '::]',
#    'word.nextisspace': False,
#    'word.nextpostag': 'NCMN',
#    'word.nextwordisdigit': False},
    features={
        'word.word': word,
        'word.isspace':word.isspace(),
        'postag':postag,
        'word.isdigit()': word.isdigit()
    }
    if i > 0:
        prevword = doc[i-1][0]
        postag1 = doc[i-1][1]
        features['word.prevword'] = prevword
        features['word.previsspace']=prevword.isspace()
        features['word.prepostag'] = postag1
        features['word.prevwordisdigit'] = prevword.isdigit()
    else:
        features['BOS'] = True # Special "Beginning of Sequence" tag
    # Features from next word
    if i < len(doc)-1:
        nextword = doc[i+1][0]
        postag1 = doc[i+1][1]
        features['word.nextword'] = nextword
        features['word.nextisspace']=nextword.isspace()
        features['word.nextpostag'] = postag1
        features['word.nextwordisdigit'] = nextword.isdigit()
    else:
        features['EOS'] = True # Special "End of Sequence" tag
    return features
```

```python
def extract_features(doc):
    return [doc2features(doc, i) for i in range(len(doc))]

def get_labels(doc):
    return [tag for (token,postag,tag) in doc]
```

```python
typ = 'sent'
path = 'final'
di = '/home/crimex/CRIMEX/Dataset/crf/'+ typ+'/'
train_path = di + 'train_'+ path
test_path = di + 'test_' + path
save_result_xls = '/home/crimex/CRIMEX/predicted_bert/' + path+ '_'+ typ+'_crf.xlsx'
filename = di+ path + '/crf_train_' + path
save_txt = di+ path + '/report_crf_train_' + path +'.txt'

print("directory path       : ", di)
print("model file           : ", filename)
print("\nTrain file           : ", train_path)
print("Test file            : ", test_path)
print("\ntxt eval result      : ", save_txt)
print("xls predicted result : ", save_result_xls)
```

```python
# with open(train_500_path+'.txt', "r") as f:
#     train_500 = [line.strip() for line in f.readlines()]
#     f.close()
    
# with open(train_1000_path+'.txt', "r") as f:
#     train_1000 = [line.strip() for line in f.readlines()]
#     f.close()
    
with open(train_path+'.txt', "r") as f:
    train = [line.strip() for line in f.readlines()]
    f.close()
    
with open(test_path+'.txt', "r") as f:
    test = [line.strip() for line in f.readlines()]
    f.close()
```

```python
# print(len(train_500))
# print(len(train_1000))
print(len(train))
print(len(test))
```

```python
import ast

first = True
datatofile = []
train_raw = []
test_raw = []

for tra in train:
    datatofile.append(list(ast.literal_eval(tra)))
    train_raw.append(list(ast.literal_eval(tra)))
    

for tes in test:
    datatofile.append(list(ast.literal_eval(tes)))
    test_raw.append(list(ast.literal_eval(tes)))
```

```python
X = [extract_features(doc) for doc in train_raw] # เอา คำ แยกออกมา
y = [get_labels(doc) for doc in train_raw] # เอา tag แยกออกมา

X_test = [extract_features(doc) for doc in test_raw] # เอา คำ แยกออกมา
y_test = [get_labels(doc) for doc in test_raw] # เอา tag แยกออกมา
```

```python
print(len(X))
print(len(X_test))
```

```python
%%time
crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    c1=0.1,
    c2=0.1,
    max_iterations=500,
    all_possible_transitions=True,
    model_filename=filename+"-pos.model" # ตั้งชื่อโมเดล
)

crf.fit(X, y); # train
```

```python
y_pred = crf.predict(X_test)
```

```python
import sklearn.metrics
import numpy
from itertools import chain

flatten_true_predictions= list(chain.from_iterable(y_pred))
flatten_true_labels= list(chain.from_iterable(y_test))
# printing result
r = sklearn.metrics.confusion_matrix(flatten_true_labels, flatten_true_predictions)
r
cm = r.astype('float') / r.sum(axis=1)[:, np.newaxis]
# cm
# cm.diagonal()
```

```python
from sklearn.metrics import matthews_corrcoef, confusion_matrix
tag_map = {'B-Criminal': 'Criminal',
           'I-Criminal': 'Criminal',
           'B-Action': 'Action',
           'I-Action': 'Action',
           'B-Location': 'Location',
           'I-Location': 'Location',
           'B-Item': 'Item',
           'I-Item': 'Item',
           'B-Victim': 'Victim',
           'I-Victim': 'Victim',
           'B-worth': 'Worth',
           'I-worth': 'Worth',
           'B-Datetime': 'Datetime',
           'I-Datetime': 'Datetime',
           'B-Enforcement': 'Enforcement',
           'I-Enforcement': 'Enforcement',
           'B-Rootcause': 'Rootcause',
           'I-Rootcause': 'Rootcause',
           'B-Trigger': 'Trigger',
           'I-Trigger': 'Trigger',
           'O': 'O'}

# Convert the original tags to merged tags
merged_true_labels = [tag_map[tag] for tag in flatten_true_labels]
merged_true_predictions = [tag_map[tag] for tag in flatten_true_predictions]

ny_true = np.array(merged_true_labels)
ny_pred = np.array(merged_true_predictions)

# mcc report (str)
mcc_report = "+--------------+-----------------------+\n"
mcc_report += f"| MCC report \n"
mcc_report += "+--------------+-----------------------+\n"

# Calculate MCC for all labels
mcc_all = matthews_corrcoef(ny_true, ny_pred)
# print("MCC (all labels):", mcc_all)

# Calculate MCC for each label
for label in np.unique(ny_true):
    y_true_label = np.where(ny_true == label, 1, 0)
    y_pred_label = np.where(ny_pred == label, 1, 0)
    mcc_label = matthews_corrcoef(y_true_label, y_pred_label)
#     print(f"MCC (label {label}):", mcc_label)
    mcc_report += f"| {label:<12} |  {mcc_label}\n"
    
mcc_report += "+---------------+-----------------------+\n"
mcc_report += f"| {'Overall':<12} |  {mcc_all}\n"
mcc_report += "+---------------+-----------------------+\n"
print(mcc_report)
```

```python
from sklearn.metrics import accuracy_score
accuracy_all = dict()
for label in np.unique(ny_true):
    y_true_label = np.where(ny_true == label, 1, 0)
    y_pred_label = np.where(ny_pred == label, 1, 0)
    accuracy = accuracy_score(y_true_label, y_pred_label)
    accuracy_all[label] = accuracy

# print("Accuracy (all labels):", np.mean(list(accuracy_all.values())))

# acc report (str)
acc_report = "+--------------+-----------------------+\n"
acc_report += f"| ACC report \n"
acc_report += "+--------------+-----------------------+\n"

for label, accuracy in accuracy_all.items():
#     print(f"Accuracy (label {label}):", accuracy)
    acc_report += f"| {label:<12} |  {accuracy}\n"
    
acc_report += "+---------------+-----------------------+\n"
acc_report += f"| {'Overall':<12} |  {np.mean(list(accuracy_all.values()))}\n"
acc_report += "+---------------+-----------------------+\n"
print(acc_report)
```

```python
from sklearn.metrics import classification_report
report = classification_report(ny_true, ny_pred, labels=np.unique(ny_true))
print(report)
```
