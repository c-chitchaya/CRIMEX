```python
def cut_title(text):
    # cut ID, DATE part 
    pattern = '[::DESC::]'
    before, sep, after = text.partition(pattern)
    new = sep + after
    return new
```

```python
def write_file_txt(data, filepath, cut=True):
    count =0

    cri=0
    vic=0
    act=0
    itm=0
    enf=0
    root=0
    trig=0
    loc=0
    dat=0
    wor =0

    with open(filepath, 'a') as f:
        for k in range(len(data)):  #range(len(data))
            # new string 
            new = ''
            ini_list_label = data['label'][k]
            ini_list_text = data['text'][k]
            
            # skip when news has no label
            if len(ini_list_label) == 0:
                # print(k, " : ", len(ini_list_label))
                continue; 

            # count no of news  
            count += 1   
            # position of the current index - new crime start with 1 
            cur_in = 0; 
            
            # loop for labeling  
            for i in ini_list_label :
                # print(i)
                if i[2] in ['Criminal', 'Victim', 'Enforcement', 'Item', 'Location', 'Datetime']: 
                # if i[2] != '': 
                    # print("GOT IT!  --  ", i[2])
                    # text before label symbol
                    new += str(ini_list_text[cur_in:int(i[0])])

                    # label name 
                    new += str(('[['+ i[2] + ']]' + str(ini_list_text[int(i[0]):int(i[1])]) + '[[/'+ i[2] + ']]'))

                    # update position 
                    cur_in = int(i[1]); 

                    if i[2] == 'Criminal': 
                      cri+=1
                    if i[2] == 'Victim': 
                      vic+=1
                    if i[2] == 'Enforcement': 
                      enf+=1
                    if i[2] == 'Item': 
                      itm+=1
                    if i[2] == 'Location': 
                      loc+=1
                    if i[2] == 'Datetime': 
                      dat+=1
                    if i[2] == 'Rootcause': 
                      root+=1
                    if i[2] == 'Trigger': 
                      trig+=1
                    if i[2] == 'worth': 
                      wor+=1
                    if i[2] == 'Action': 
                      act+=1

        
            # another text 
            new += str(ini_list_text[cur_in:len(ini_list_text)])
            # remove \n
            new = new.replace("\n", "")

            if cut == True:
                new = cut_title(new)

            new=re.sub(' +', ' ',new) # remove multiple whitespace
            # print(new)

            # write "new" to text file 
            f.write(new)
            f.write("\n") # new line 
    
    # close file   
    f.close();

    # print('no of news: ', count)
    # print('location -> ', filepath)
    
    return count, cri, vic, act, itm, enf, root, trig, loc, dat, wor
```
